```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import math


# In[59]:



df=pd.read_csv("data.csv")

X=df["Annual Income (k$)"]
Y=df["Spending Score (1-100)"]


data=list(zip(X,Y))


c1=random.choice(data)
c2=random.choice(data)
c3=random.choice(data)
c4=random.choice(data)
c5=random.choice(data)


for iteration in range(1000):
  d1=[]
  d2=[]
  d3=[]
  d4=[]
  d5=[]

  cluster=[]

  v=(len(data))
  distance_list=[]
  for i in  range(v):
    distance=math.dist(data[i],c1)
    d1.append(distance)
    distance=math.dist(data[i],c2)
    d2.append(distance)
    distance=math.dist(data[i],c3)
    d3.append(distance)
    distance=math.dist(data[i],c4)
    d4.append(distance)
    distance=math.dist(data[i],c5)
    d5.append(distance)




  clusters=[]
  for i in range(v):
    temp_list=[]

    temp_list.append(d1[i])
    temp_list.append(d2[i])
    temp_list.append(d3[i])
    temp_list.append(d4[i])
    temp_list.append(d5[i])

    min_index=temp_list.index(min(temp_list))

    clusters.append(min_index)


  cluster0=[]
  cluster1=[]
  cluster2=[]
  cluster3=[]
  cluster4=[]

  l=len(clusters)

  for i in range(l):


    if clusters[i]==0:
        cluster0.append(data[i])


    if clusters[i]==1:
        cluster1.append(data[i])
    if clusters[i]==2:
        cluster2.append(data[i])
    if clusters[i]==3:
      cluster3.append(data[i])
    if clusters[i]==4:
        cluster4.append(data[i])

# Replace the lines where you calculate the means of clusters
  c1 = (np.mean([point[0] for point in cluster0]), np.mean([point[1] for point in cluster0]))
  c2 = (np.mean([point[0] for point in cluster1]), np.mean([point[1] for point in cluster1]))
  c3 = (np.mean([point[0] for point in cluster2]), np.mean([point[1] for point in cluster2]))
  c4 = (np.mean([point[0] for point in cluster3]), np.mean([point[1] for point in cluster3]))
  c5 = (np.mean([point[0] for point in cluster4]), np.mean([point[1] for point in cluster4]))


```

```python
# Plotting the clusters and centroids
plt.figure(figsize=(8, 6))
x=[point[0]for point in cluster0]
y=[point[1]for point in cluster0]
plt.scatter(x,y,color='red')
x=[point[0]for point in cluster1]
y=[point[1]for point in cluster1]
plt.scatter(x,y,color='blue')
x=[point[0]for point in cluster2]
y=[point[1]for point in cluster2]
plt.scatter(x,y,color='purple')
x=[point[0]for point in cluster3]
y=[point[1]for point in cluster3]
plt.scatter(x,y,color='green')
x=[point[0]for point in cluster4]
y=[point[1]for point in cluster4]
plt.scatter(x,y,color='black')

plt.scatter(c1[0],c1[1],color='red',marker='X')

plt.scatter(c2[0],c2[1],color='red',marker='X')

plt.scatter(c3[0],c3[1],color='red',marker='X')
plt.scatter(c4[0],c4[1],color='red',marker='X')

plt.scatter(c5[0],c5[1],color='red',marker='X')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('K-Means Clustering')
plt.legend()
plt.grid(True)
plt.show()
```
